```python
# import everything here
import os
import pandas as pd
import shutil
```

```python
def file_separator(csv_list, base_dir, output_dir):
    id_df = pd.read_csv(csv_list)
    
    for index, row in id_df.iterrows():
        data_id = row['Participant_ID']
        
        source_file = os.path.join(base_dir, f'{data_id}_P~', f'{data_id}_P')


        if os.path.isdir(source_file):
            dest_folder = os.path.join(output_dir, str(data_id))
            

            if not os.path.exists(dest_folder):
                os.makedirs(dest_folder)
            
            shutil.move(source_file, dest_folder)
            print(f'Moved {data_id} folder to output directory')
```

```python
train = "labels/train_split.csv"
test = "labels/test_split.csv"

base_dir = "D:\Developer\DepressionDetection\depresdata"
train_output_dir = "train"
test_output_dir = "test"
```

```python
file_separator(train, base_dir, train_output_dir)
```

```python
file_separator(test, base_dir, test_output_dir)
```
